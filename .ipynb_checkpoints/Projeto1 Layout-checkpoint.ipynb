## Como a renda mensal per capita e a quantidade de moradores em um domicílio afetam na   segurança alimentar dos jovens na regiao nordeste?

___
## Introdução

   Nesse projeto tenho como objetivo analisar a seguraça alimentar por meio das PNADS do IBGE. Mais especificamente irei analisar na PNAD de 2009 os casos de (in)segurança alimentar entre os jovens (menores de 18 anos) no nordeste e sua relação com a renda mensal per capita e o numero de moradores no domicílio.


___
## Análise e Resultados

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
dados2009 = pd.read_csv('projeto1/DOM2009.csv')
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#filtrando as colunas:
dados09 = dados2009.loc[:,["UF","V0105","V4621","V4623","V2124","V2126","V2132","V2134"]]
#filtrando para apenas domicilios que possuam menores de idade:
dados09 = dados09[dados09.V4623<=4]
#ajeitando a veriaveis de renda mensal:
dados09 = dados09[dados09.V4621 != 999999999999]
#variaveis qualitativas:
dados09.V2124 = dados09.V2124.astype("category")
dados09.V2124.cat.categories = (['Sim','Não'])
dados09.V2126 = dados09.V2126.astype("category")
dados09.V2126.cat.categories = (['Sim','Não'])
dados09.V2132 = dados09.V2132.astype("category")
dados09.V2132.cat.categories = (['Sim','Não'])
dados09.V2134 = dados09.V2134.astype("category")
dados09.V2134.cat.categories = (['Sim','Não'])

dados09.V4623 = dados09.V4623.astype("category")
dados09.V4623.cat.categories = (['boa','leve','moderada','grave'])
#renomeando as variaveis:
dados09 = dados09.rename(columns = {"V4621":"renda",'V0105':'moradores',"V4623":"seg_alimentar" , "V2124":"deixou_de_ter_alimentos_saudaveis" , "V2126":"nao_comeu_suficiente","V2132":"sentiu_fome","V2134":"menos_de_uma_refeicao"})
#utilizando apenas os estados do nordeste:
dados09nordeste = dados09[(dados09.UF >= 21) & (dados09.UF <= 29)]
#listando as variaveis:
print('lista de variáveis utilizadas:')
list(dados09nordeste)
```

### Gráfico da segurança alimentar dos jovens( menores 18) no nordeste

```python
#grafico de seguranca alimentar
gsa09 = dados09nordeste.seg_alimentar.value_counts(True)*100

fig = plt.figure(figsize=(12,5))

plt.subplot(121)
plot = gsa09.plot(kind ='bar', color = ('green','yellow','orange','red'))
plt.title("segurança alimentar em jovens no nordeste")
plt.ylabel("porcentagem")

plt.subplot(122)
plot = gsa09.plot(kind ='pie',colors = ('green','yellow','orange','red'), legend = False, )
plt.title("segurança alimentar em jovens no nordeste")
```

    boa = boa segurança alimentar;
    leve = insegurança alimentar leve;
    moderada = insegurança alimentar moderada;
    grave = insegurança alimentar grave.

### Definindo a insegurança alimentar
essa parte tem como o objetivo de mostrar a diferença entre os tres tipos de insegurança alimentar (leve, moderada, grave) a partir de 4 variaveis:
- jovens que sentiram fome;
- jovens que fizeram menos de uma refeiçao por dia;
- jovens que nao comeram o suficiente;
- jovens que deixaram de ter alimentos saudaveis.

```python

ctsegalimentar = pd.crosstab(dados09nordeste.seg_alimentar,dados09nordeste.sentiu_fome)
ctsegalimentar1 = pd.crosstab(dados09nordeste.seg_alimentar,dados09nordeste.menos_de_uma_refeicao)
ctsegalimentar2 = pd.crosstab(dados09nordeste.seg_alimentar,dados09nordeste.nao_comeu_suficiente)
ctsegalimentar3 = pd.crosstab(dados09nordeste.seg_alimentar,dados09nordeste.deixou_de_ter_alimentos_saudaveis)
ctsegalimentar.plot(kind = 'bar')
plt.xlabel('inseguranca alimentar')
plt.title('jovens com insegurança alimentar que\n sentiram fome')
ctsegalimentar1.plot(kind = 'bar')
plt.xlabel('inseguranca alimentar')
plt.title('jovens com insegurança alimentar que\n tiveram uma ou menos refeição no dia')

ctsegalimentar2.plot(kind = 'bar')
plt.xlabel('inseguranca alimentar')
plt.title('jovens com insegurança alimentar que\n nao comeram o suficiente')

ctsegalimentar3.plot(kind = 'bar')
plt.xlabel('inseguranca alimentar')
plt.title('jovens com insegurança alimentar que\n deixaram de ter alimentos saudaveis')
```

### Variáveis quantitativas
#### Renda mensal per capita e total de moradores por domicílio

```python
#histograma da renda
faixa = range(0,3000,100)
plotrenda = dados09nordeste.renda.plot.hist(bins=faixa,title='histograma de renda mensal domiciliar per capita do nordeste',normed = 1,figsize=(10,5))
plt.xlabel("rendimento domiciliar mensal per capita")
```

```python
#histograma dos moradores
plotmoradores = dados09nordeste.moradores.plot.hist(bins=range(0,20,2),title='histograma de moradores por dmicilio do nordeste',normed = 1,figsize=(10,5))
plt.xlabel('total de moradores por domicilio')
from matplotlib.ticker import MaxNLocator
plotmoradores.xaxis.set_major_locator(MaxNLocator(integer=True))
```

```python
#juntando as duas variaveis quantitativas:
plt.scatter(dados09nordeste.moradores, dados09nordeste.renda)
plt.axis([0,15,0,8000])
plt.ylabel('renda')
plt.xlabel('total de moradores')
plt.title('Scatter\nrenda domiciliar X moradores no domicilio')
```

Como observado no grafico acima, o total de moradores está relacionado com a renda, já que quanto maior o numero de indivíduos no domicilio representa uma renda mensal per capita menor para a maioria dos casos

### Relacionando a segurança alimentar com o total de moradores por domicilio
A seguir estão apresentados dois gráficos:
    Um com o total de moradores de 0 à 16 agrupando de 2 em 2 ou seja (0-2,2-4,4,6...) e o segundo grafico como se fosse o zoom do primeiro mostrando apenas os domicilios de 8 à 16 moradores

```python
moradores = pd.cut(dados09nordeste.moradores, bins=range(0,17,2))
moradores2 = pd.cut(dados09nordeste.moradores, bins=range(8,17,2))
ctnordeste = pd.crosstab(moradores, dados09.seg_alimentar)
ct2nordeste = pd.crosstab(moradores2, dados09.seg_alimentar)

ctnordeste.plot(kind = 'bar', figsize = (10,5), color = ('green','yellow','orange','red'))
plt.title('segurança alimentar por moradores no domicilio [0-16]')

ct2nordeste.plot(kind = 'bar', figsize = (10,5), color = ('green','yellow','orange','red'))
plt.title('segurança alimentar por moradores no domicilio [8-16]')
```

Observando o gráfico pode-se perceber como a segurança alimentar ('boa') diminui comparado à inseguranca alimentar ('leve','moderado','grave') conforme o numero de moradores no domicilio aumenta

### Gráficos conclusivos
Abaixo estão os gráficos realacionando as 4 faixas de segurança alimentar (segurança alimentar boa, insegurança alimentar leve, insegurança moderada e insegurança alimentar grave) com os dados quantitativos de renda mensal domiciliar per capita e quantidade de moradores por domicílio

```python
dadosboa = dados09nordeste[dados09nordeste.seg_alimentar == 'boa']
dadosleve = dados09nordeste[dados09nordeste.seg_alimentar == 'leve']
dadosmoderado = dados09nordeste[dados09nordeste.seg_alimentar == 'moderada']
dadosgrave = dados09nordeste[dados09nordeste.seg_alimentar == 'grave']
faixa = range(0,3000,100)
faixa2 = range(0,17,2)
fig = plt.figure(figsize=(16,16))

plt.subplot(421)
plot = dadosboa.renda.plot.hist(bins=faixa,title='[BOA] renda',normed = 1, color = 'green')
plt.xlabel("rendimento domiciliar mensal per capita")

plt.subplot(422)
plot = dadosboa.moradores.plot.hist(bins=faixa2,title='[BOA] moradores',normed = 1, color = 'green')
plt.xlabel("total de moradores o domicilio")

plt.subplot(423)
plot = dadosleve.renda.plot.hist(bins=faixa,title='[LEVE] renda',normed = 1, color = 'yellow')
plt.xlabel("rendimento domiciliar mensal per capita")

plt.subplot(424)
plot = dadosleve.moradores.plot.hist(bins=faixa2,title='[LEVE] moradores',normed = 1, color = 'yellow')
plt.xlabel("total de moradores o domicilio")

plt.subplot(425)
plot = dadosmoderado.renda.plot.hist(bins=faixa,title='[MODERADO] renda',normed = 1, color = 'orange')
plt.xlabel("rendimento domiciliar mensal per capita")


plt.subplot(426)
plot = dadosmoderado.moradores.plot.hist(bins=faixa2,title='[MODERADO] moradores',normed = 1, color = 'orange')
plt.xlabel("total de moradores o domicilio")

plt.subplot(427)
plot = dadosgrave.renda.plot.hist(bins=faixa,title='[GRAVE] renda',normed = 1, color = 'red')
plt.xlabel("rendimento domiciliar mensal per capita")

plt.subplot(428)
plot = dadosgrave.moradores.plot.hist(bins=faixa2,title='[GRAVE] moradores',normed = 1, color = 'red')
plt.xlabel("total de moradores o domicilio")
plt.tight_layout()
plt.show()
```

___
## Conclusão 

Como observado nos histogramas, há um clara relação entre as três variáveis. Conforme baixamos o nível de segurança alimentar (de boa para grave) percebe-se que a renda mensal domiciliar per capita dimínui e o total de moradores no domíclio tende a crescer.
Pode-se concluir que quanto menor a renda e maior o numero de moradores há mais chance de haver algum tipo de insegurança alimentar
